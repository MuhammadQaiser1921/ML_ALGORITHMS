```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/sample_data/Salary_Data.csv")
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
sns.lmplot(x="Age",y="Salary",data=dataset)
plt.show()
```

```python
# prompt: create graph to check linear characteristics of dataset


sns.lmplot(x='YearsExperience',y='Salary',data=dataset)
plt.show()
```

```python
x = dataset[["YearsExperience", "Age"]]
y = dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict([[1.1,21.0]])
```

```python
model.score(x_test,y_test)*100
```
